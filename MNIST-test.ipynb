```python
import torch
import torch.nn as nn
import torchvision
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
```

```python
transforme = transforms.Compose([transforms.ToTensor() #turns the pixels into tensors
                   ,transforms.Normalize(0.1445, 0.3204)])#Normalizes the input values
```

```python
train_data = datasets.MNIST(root='data',download=True, train=True, transform= transforme)
test_data = datasets.MNIST(root='data',download=True, train=False, transform= transforme)
```

```python
train_loader = DataLoader(train_data, 42, shuffle= True) #creates our dataset with a shuffle to prevent order of classification categories, each batch has 35 inputs that we get the mean of
test_loader = DataLoader(test_data, 1000, shuffle= False)
```

```python
x,y = next(iter(train_loader)) #first batch of inputs (pixels) and outpts (labels)
x[0].shape
```

```python
flatten = nn.Flatten() #we will flatten the image to pass it through our network (we will make it 1D instead of 2D to apply matrix multiplications)
model_k = nn.Sequential(
            nn.Linear(784, 150),
            nn.ReLU(),
            nn.Linear(150, 10)
        )

    
 #flatten the input
model_k(flatten(x[0])) #pass it through our NN that has random weights and biases
```

```python
loss_fn = nn.CrossEntropyLoss()
optim = torch.optim.Adam(model_k.parameters(),lr=0.0001)
```

```python
import numpy as np
flatten(x[0]).mean() #get the mean and std for the normalize function above
flatten(x[0]).std() #std = 0.3204, mean = 0.14
```

```python
model_k.train(True)
batch = 0
for idx, idy in train_loader:
    print(batch)
    idx = flatten(idx)
    outputs = model_k(idx)
    print(outputs.dtype,outputs.shape)
    loss = loss_fn(outputs,idy)
    
    optim.zero_grad()
    print(loss)
    loss.backward()
    optim.step()
```
